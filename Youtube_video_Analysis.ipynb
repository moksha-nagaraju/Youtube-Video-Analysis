```python
#Removing unnecessary columns for further analysis
df.drop(["channelId","thumbnail_link","description","tags"], inplace=True, axis=1)
```

```python
df_maxViews = df.sort_values('view_count', ascending=False).drop_duplicates(['video_id'])
```

```python
df_maxViews.sample(5)
```

```python
df_minViews = df.sort_values('view_count', ascending=False).drop_duplicates(['video_id'])
df_minViews.sample(5)
```

```python
df_Channel_Analysis = df_maxViews.groupby('channelTitle')["view_count"].sum().reset_index()
df_Like = df_maxViews.groupby('channelTitle')["likes"].sum().reset_index()
df_Dislike = df_maxViews.groupby('channelTitle')["dislikes"].sum().reset_index()
df_Comments = df_maxViews.groupby('channelTitle')["comment_count"].sum().reset_index()
df_Channel_Analysis["Likes"] = df_Like.likes
df_Channel_Analysis["Dislikes"] = df_Dislike.dislikes
df_Channel_Analysis
```

```python
#  Most Trending YouTube Categories 
plt.figure(figsize=(10, 6))
sns.countplot(y=df["categoryId"], order=df["categoryId"].value_counts().index, palette="coolwarm")
plt.title("Top Trending YouTube Categories in India")
plt.xlabel("Count")
plt.ylabel("Category ID")
plt.show()
```

```python
# Engagement Analysis: Views vs. Likes ----
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="view_count", y="likes", alpha=0.5)
plt.title("Views vs. Likes")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Views (log scale)")
plt.ylabel("Likes (log scale)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="likes", y="dislikes", alpha=0.5, color="red")
plt.title("Likes vs. Dislikes")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Likes (log scale)")
plt.ylabel("Dislikes (log scale)")
plt.show()
```

```python
#  How Many Videos Have Disabled Comments? ----
plt.figure(figsize=(6, 5  ))
df["comments_disabled"].value_counts().plot(kind="bar", color=["green", "red"])
plt.title("How Many Videos Have Disabled Comments?")
plt.xticks(ticks=[0, 1], labels=["Comments Enabled", "Comments Disabled"], rotation=0)
plt.ylabel("Count")
plt.show()
```

```python
# Most Popular YouTube Channels (Top 10 by Trending Videos) ----
plt.figure(figsize=(10, 6))
top_channels = df["channelTitle"].value_counts().head(10)
sns.barplot(y=top_channels.index, x=top_channels.values, palette="coolwarm")
plt.title("Top 10 Most Trending YouTube Channels in India")
plt.xlabel("Number of Trending Videos")
plt.ylabel("Channel Name")
plt.show()
```

```python
#  Insights
print("\n Top 5 Most Popular YouTube Channels:")
print(df["channelTitle"].value_counts().head(5))
```

```python
# Pie Chart for Videos with Disabled Ratings ----
plt.figure(figsize=(10, 6))
df["ratings_disabled"].value_counts().plot.pie(autopct="%1.1f%%", colors=["blue", "orange"], startangle=90)
plt.title("Percentage of Videos with Ratings Disabled")
plt.xticks(ticks=[])  # Hide x-ticks
plt.ylabel("")  # Hide y-label
plt.show()
```
